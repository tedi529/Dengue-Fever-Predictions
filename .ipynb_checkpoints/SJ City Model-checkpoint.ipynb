# SJ City - Model for Prediction of Dengue Fever

```python
# Import dependencies
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
import matplotlib.pyplot as plt

import os
import pandas as pd
import numpy as np
import seaborn as sb

from numpy import arange
from numpy import absolute
from numpy import mean
from numpy import std

from scipy import stats
```

## Data Import and Pre-Processing

```python
# Import features training dataset
features_training_data = os.path.join("Data", "dengue_features_train.csv")
features_training_df = pd.read_csv(features_training_data)
features_training_df.head()
features_training_df
```

**There are 1456 training observations.**

```python
# Import features testing dataset
features_testing_data = os.path.join("Data", "dengue_features_test.csv")
features_testing_df = pd.read_csv(features_testing_data)
features_testing_df
```

**There are 416 test observations.**

```python
# Import training labels dataset, leave only total_cases
labels_training_data = os.path.join("Data", "dengue_labels_train.csv")
labels_training = pd.read_csv(labels_training_data)
labels_training_df = labels_training.drop(columns=['city', 'year', 'weekofyear'])
labels_training_df
```

### Process Training Data

```python
# Check duplicate columns
features_training_df['precipitation_amt_mm'].equals(features_training_df['reanalysis_sat_precip_amt_mm'])
```

```python
# Drop unneccesary and duplicate features 
features_training_processing = features_training_df.drop(columns=['year','weekofyear','week_start_date','precipitation_amt_mm'])
features_training_processing

# Combine data
combined_df = pd.concat([features_training_processing, labels_training_df], axis=1, join='inner') 

# Keep only SJ city data
combined_df_sj = combined_df.loc[combined_df['city'] == 'sj']
combined_df_sj
```

## Data Exploration & Feature Selection

```python
# Examine pearson correlation matrix to see if some variables should be excluded from analysis
corr_matrix = combined_df_sj.corr(method='pearson') #.unstack().sort_values().drop_duplicates()

# Create heatmap for easier visualization
plt.figure(figsize = (25,15))
lower = np.triu(corr_matrix)
sb.heatmap(corr_matrix,cmap=sb.diverging_palette(220, 10, as_cmap=True), annot=True, square=True, fmt='.3f', mask=lower)

# Customary to keep variables with correlation coefficients smaller than absolute value of 0.8
```

```python
# Examine scatterplot of features
sb.pairplot(data=combined_df_sj, 
                 x_vars=['total_cases'],
                 y_vars=combined_df.drop(['total_cases','city'], axis=1),
                 size=3)

plt.show()
```

```python
# Keep feature with higher correlation coefficients
kept_features_df = combined_df_sj[['city','reanalysis_air_temp_k','reanalysis_avg_temp_k','reanalysis_dew_point_temp_k',
                                'reanalysis_max_air_temp_k','reanalysis_min_air_temp_k','reanalysis_specific_humidity_g_per_kg',
                                'station_avg_temp_c','station_max_temp_c','station_min_temp_c','total_cases']]

# kept_features_df = combined_df_sj[['city','reanalysis_dew_point_temp_k','reanalysis_max_air_temp_k',
#                                    'reanalysis_specific_humidity_g_per_kg',
#                                    'station_avg_temp_c','total_cases']]

# kept_features_df = combined_df_sj.drop(columns=['ndvi_ne','ndvi_sw','reanalysis_air_temp_k',
#                                                 'reanalysis_specific_humidity_g_per_kg','station_diur_temp_rng_c',
#                                                 'reanalysis_tdtr_k'])
kept_features_df
```

```python
# Drop NaN values
kept_features_df_dropped = kept_features_df.dropna()

#  Impute to NaN values previous week's value for the same feature for each city, respectively
# kept_features_df_filled = kept_features_df.fillna(method='ffill')
```

```python
# Drop city column
kept_features_df_dropped.drop(columns=['city'])
# kept_features_df_filled.drop(columns=['city'])
```

### Remove Outliers

```python
# Remove rows with outliers
no_outliers_df = kept_features_df_dropped.drop(columns=['city'])[(np.abs(stats.zscore(kept_features_df_dropped.drop(columns=['city']))) < 3).all(axis=1)]
# no_outliers_df = kept_features_df_filled.drop(columns=['city'])[(np.abs(stats.zscore(kept_features_df_filled.drop(columns=['city']))) < 3).all(axis=1)]
```

# TEST MODEL

```python
# Import models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import LassoCV
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import ElasticNetCV
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error
```

```python
# Assign training dataset into X(features) and y(label)
X = no_outliers_df.drop(['total_cases'], axis=1)
y = no_outliers_df['total_cases'].values.reshape(-1,1)
print(X.shape, y.shape)

X_train, X_eval, y_train, y_eval = train_test_split(X, y, test_size=0.33, random_state=42)
X_train.shape, X_eval.shape
print(X_train.shape, X_eval.shape)
```

```python
models = [LinearRegression(),
          Lasso(alpha=0.04),
          ElasticNet(alpha=0.01,l1_ratio=0.49),
          Ridge(alpha=1.0),
          GradientBoostingRegressor()]
```

```python
# Fit (train) baseline model using training data
for model in models:
    model.fit(X_train, y_train)

    # Evaluate baseline model
    yhat = model.predict(X_eval)

    # Evaluate predictions
    mae = mean_absolute_error(y_eval, yhat)
    print(f'Dropped Features {model} MAE: %.3f' % mae)
```

## Submission Preparation

```python
# Split data into cities
testing_df_sj = features_testing_df.loc[features_testing_df['city'] == 'sj']

# Impute to NaN values previous week's value for the same feature for each city, respectively
testing_df_sj.fillna(method='ffill', inplace=True)
```

## Assign Model and Make Predictions

```python
# Train model on full training dataset
model = Lasso(alpha=0.04)
model.fit(X, y)
```

```python
# Make predictions on test set
test_predictions_raw = model.predict(testing_df_sj)
test_predictions_raw[test_predictions_raw < 0] = 0
test_predictions_sj = test_predictions_raw.astype(int)
```
